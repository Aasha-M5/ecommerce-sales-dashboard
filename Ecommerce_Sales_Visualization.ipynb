```python
import pandas as pd
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
```

Display Dataset


```python
df.head()
```

Check Number of Rows and Columns

```python
df.shape
```

View Column Names

```python
df.columns
```

Understand Dataset Information

```python
df.info()
```

Statistical Summary

```python
df.describe()
```

Check Missing Values

```python
df.isnull().sum()
```

Remove Missing Values

```python
df = df.dropna()
```

df.duplicated().sum()

```python
df.duplicated().sum()
```

Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

Check Data Types

```python
df.dtypes
```

 Convert Date Column

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

Create Year Column

```python
df['Year'] = df['Order Date'].dt.year
```

 Create Month Column

```python
df['Month'] = df['Order Date'].dt.month_name()
```

Verify Cleaned Dataset

```python
df.head()
```

Total Sales

```python
total_sales = df['Sales'].sum()

print("Total Sales:", total_sales)
```

Total Profit

```python
total_profit = df['Profit'].sum()

print("Total Profit:", total_profit)
```

Total Orders

```python
total_orders = df['Order ID'].nunique()

print("Total Orders:", total_orders)
```

Category-wise Sales


```python
category_sales = df.groupby('Category')['Sales'].sum()

print(category_sales)
```

Category-wise Profit

```python
category_profit = df.groupby('Category')['Profit'].sum()

print(category_profit)
```

Region-wise Sales

```python
region_sales = df.groupby('Region')['Sales'].sum()

print(region_sales)
```

 Region-wise Profit

```python
region_profit = df.groupby('Region')['Profit'].sum()

print(region_profit)
```

Top 10 Products by Sales

```python
top_products = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

print(top_products.head(10))
```

Monthly Sales Trend

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

print(monthly_sales)
```

Average Sales

```python
average_sales = df['Sales'].mean()

print("Average Sales:", average_sales)
```

 Import Visualization Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Bar Chart Category-wise Sales

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')

plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

Pie Chart
Region-wise Sales Percentage

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.pie(region_sales,
        labels=region_sales.index,
        autopct='%1.1f%%')

plt.title("Region-wise Sales Percentage")

plt.show()
```

Line Chart
Monthly Sales Trend


```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

Histogram
Sales Distribution


```python
plt.hist(df['Sales'], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

Scatter Plot
 Sales vs Profit


```python
plt.scatter(df['Sales'], df['Profit'])

plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

Heatmap
Correlation Heatmap


```python
numeric_df = df.select_dtypes(include=['number'])

correlation = numeric_df.corr()

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
